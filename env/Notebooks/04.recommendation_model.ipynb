```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

# Load the preprocessed data
df = pd.read_csv('../artifacts/preprocessed_with_sentiments.csv')
user_product_matrix = pd.read_csv('../artifacts/user_product_matrix.csv', index_col=0)
product_features = pd.read_csv('../artifacts/product_features.csv', index_col=0)

print("Data loaded successfully!")
print(f"Reviews shape: {df.shape}")
print(f"User-Product matrix shape: {user_product_matrix.shape}")
print(f"Product features shape: {product_features.shape}")

# Use aspect scores for content-based recommendations
aspect_cols = [col for col in df.columns if '_score' in col]
X = df[aspect_cols].fillna(0)

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate similarity
similarity_matrix = cosine_similarity(X_scaled)

print(f"\nSimilarity matrix shape: {similarity_matrix.shape}")
print("Content-based model initialized!")
```
